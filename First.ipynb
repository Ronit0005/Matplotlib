```python
import numpy as np
from matplotlib import pyplot as plt
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Data")
plt.grid()
plt.show()
```

```python
plt.plot(x,y)
plt.grid(axis='x')
plt.show()
```

```python
plt.plot(x,y)
plt.grid(axis='y')
plt.show()
```

```python
plt.plot(x,y)
plt.grid(axis='both',c='r',lw='2',ls=":")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot([1,2,3,4,5],[10,9,8,7,6])
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.plot(x,y)
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

First subplot then plot

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.title("Sales")
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot([10,20,30,40])
plt.title("Income")
plt.suptitle("My Shop")
plt.show()
```

Pie chart 

```python
x=np.array([35,25,25,15])
plt.pie(x)
plt.show()
```

```python
x=np.array([35,25,25,15])
my_label=np.array(['apple','banana','cheries','dates'])
plt.pie(x,labels=my_label)
plt.show()
```

```python
x=np.array([35,25,25,15])
my_label=np.array(['apple','banana','cheries','dates'])
plt.pie(x,labels=my_label,startangle=90)
plt.show()
```

```python
x=np.array([35,25,25,15])
my_label=np.array(["Apple","Banana","Cherries","Dates"])
my_explode=np.array([0.1,0,0,0])
plt.pie(x,labels=my_label,explode=my_explode)
plt.show()
```

```python
plt.pie(x,labels=my_label,explode=my_explode,shadow=True)
plt.show()
```

```python
plt.pie(x,labels=my_label,explode=my_explode,colors=["blue","white","red","yellow"])
plt.legend(title="Fruis")
plt.show()
```
