```python
plt.plot(y,ls=':',lw=2,c='r')
plt.xlabel("Average Pulse")
plt.ylabel("Colorie Burnage")
plt.title("Data")
plt.show()
```

```python
font1={'family':'serif','color':'red','size':10}
font2={'family':'serif','color':'blue','size':10}
font3={'family':'serif','color':'black','size':10}
plt.plot(x,y)
plt.xlabel('Average Pulse',fontdict=font1)
plt.ylabel('Colorie Burnage',fontdict=font2)
plt.title('Data',fontdict=font3)
plt.show()
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
plt.plot(x,y)
plt.grid(axis='x',c='r',ls='-.',lw=1)
plt.grid()
```

```python
plt.subplot(1,2,1)
plt.plot([10,9,8,7,6],ls=':',c='r')
plt.xlabel("1st x-axis")
plt.ylabel("1st y-axis")
plt.title('1st graph')
plt.subplot(1,2,2)
plt.plot([1,2,3,4,5],ls=':',c='r')
plt.xlabel("2nd x-axis")
plt.ylabel("2nd y-axis")
plt.title('2nd graph')
plt.show()
```

Scatter plot

```python
x=np.array([1,2,3,4,5])
y=np.array([10,9,8,7,6])
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,c='r')
plt.show()
```

For displaying all points with different color .

```python
colors=np.array(['red','cyan','blue','black','green'])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
sizes=np.array([10,20,30,40,50])
plt.scatter(x,y,s=sizes,alpha=0.1)
plt.show()
```

Bar Graph 

```python
x=np.array(['A','B','C','D'])
y=np.array([1,2,3,4])
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
plt.bar(x,y,color='black')
plt.show()
```

```python
plt.bar(x,y,width=0.5)
plt.show()
```

```python
plt.barh(x,y,height=0.3)
plt.show()
```

Histograph

```python
x=np.random.normal(180,10,300)
plt.hist(x)
plt.show()
```

Pie Chart

```python
x=np.array([35,25,25,15])
plt.pie(x)
plt.show()
```

```python
my_label=np.array(['mango','banana','guava','watermelon'])
plt.pie(x,labels=my_label)
plt.show()
```

```python
plt.pie(x,labels=my_label,startangle=90)
plt.show()
```

```python
my_explode=np.array([0.6,0,0,0.2])
plt.pie(x,labels=my_label,startangle=270,explode=my_explode)
plt.show()
```

```python
plt.pie(x,labels=my_label)
plt.legend()
plt.show()
```

```python
my_color=np.array(['red','yellow','black','blue'])
plt.pie(x,labels=my_label,colors=my_color)
plt.show()
```
